# Import Statements

```python
import keras
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.datasets import fashion_mnist
from keras.preprocessing.image import array_to_img
```

# Constants

```python
IMAGE_HEIGHT = 28
IMAGE_WIDTH = 28
VALIDATION_SIZE = 10000
NR_IMAGES = 60000
TOTAL_INPUTS = IMAGE_HEIGHT * IMAGE_WIDTH
LABEL_NAMES = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

# Gathering the data

```python
data = fashion_mnist.load_data()
```

```python
(x_train, y_train), (x_test, y_test) = data
```

```python
x_test_somethin, y_test_somethin = x_test, y_test
```

```python
for i in range(9):
    plt.imshow(x_train[i])
    plt.xlabel(LABEL_NAMES[y_train[i]], fontsize = 15)
    plt.xticks([])
    plt.yticks([])
    
    plt.show()
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(x_train.shape[0], TOTAL_INPUTS)
```

```python
x_train.shape
```

```python
x_train_processed, x_test_processed = x_train/255, x_test/255
```

```python
x_train_processed.shape
```

# Architecturing our model

```python
model_1 = Sequential([
    Dropout(0.40, seed = 67, input_shape = (TOTAL_INPUTS,)),
    Dense(units=128, activation = 'relu', name='M1_hidden_1'),
    Dropout(0.40, seed = 67),
    Dense(units = 64, activation = 'relu', name='M1_hidden_2'),
    Dropout(0.40, seed = 67),
    Dense(units = 32, activation='relu', name='M1_hidden_3'),
    Dense(units = 16, activation = 'relu', name='M1_hidden_4'),
    Dropout(0.40, seed = 67),
    Dense(units = 10, activation = 'softmax', name='M1_output')
])
model_1.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
```

```python
sample_per_batch = 1000
nr_os_iterations= 100
```

```python
def callback_kinda():
    print('Yeah Its Working')
```

```python
%%time

model_1.fit(x_train_processed, y_train,
            batch_size=sample_per_batch, epochs=nr_of_iterations, verbose=0)
```
